```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv(r'C:\Users\Dell\Desktop\ANALYTICS\Data Science Project by Python\President Height\president_heights.csv')
```

```python
data.head()
```

```python
data = pd.DataFrame(data)
```

```python
data.set_index('order', inplace=True)
```

```python
sns.histplot(data=data, x='height(cm)')
plt.xlabel('Height')
plt.ylabel('Count')
plt.title('Height Distribution')
```

```python
sns.boxplot(data=data, x='height(cm)')
```

```python
print("President Mean height is :", np.mean(data['height(cm)']))
print("President STD height is :", np.std(data['height(cm)']))
print("President Max height is :", np.max(data['height(cm)']))
print("President Min height is :", np.min(data['height(cm)']))
print("President 25th quartile :", np.percentile(data['height(cm)'],25))
print("President 75th quartile :", np.percentile(data['height(cm)'],75))
```
